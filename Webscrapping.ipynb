```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
URL = "https://lhr.nu.edu.pk/faculty/"
r = requests.get(URL)  
print(r) 
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify()) 
```

```python
f = soup.findAll('div', class_='facultyCard')
len(f)
```

```python
id_=[]
name=[]
email=[]
desig=[]
img=[]
profile=[]
higher_edu=[]
ext=[]
department=[]
hec=[]
count=0
for i in range(len(f)):
    faculty = f[i]
    try:
        id_.append(int(faculty.find('a',class_='faculty-link')['href'].split('/')[-1]))
    except:
        id_.append(-1)
    print(faculty.find('h5', class_='text-center').text)
    count=count+1
    print(count)
    name.append(faculty.find('h5', class_='text-center').text)
    desig.append(faculty.find('p',class_='small').text.strip().split('\n')[0])
    try:
        if(faculty.find('p',class_='small').text.strip().split('\n')[2]):
            hec.append("True")
    except:
        hec.append('none')
    img.append('https://lhr.nu.edu.pk'+faculty.find('img', class_='card-img-top')['src'])
    email.append(faculty.find('p', class_='mb-0').text)
    department.append(faculty.find('a',class_='faculty-link')['href'].split('/')[-3])
    url1= "https://lhr.nu.edu.pk"+(faculty.find('a',class_='faculty-link')['href'])
    print(url1)
    r1 = requests.get(url1) 
    print(r1) 
    soup1 = BeautifulSoup(r1.content, 'html.parser') 
    faculty1 = soup1.find('div', class_='facultyCard')
    try:
        ext.append(faculty1.find('span', class_='small').text.split(':')[-1])
    except:
        ext.append('none')
    try:
        high_edu=soup1.find('div',class_='col-lg-8')
        higher_edu.append((high_edu.find('li').text))
    except:
        higher_edu.append('none')
```

```python
df1 = pd.DataFrame({
                  'ID':id_,
                  'Name':name,
                  'Designation':desig,
                  'HEC approved':hec,
                 'Education':higher_edu,
                  'Email':email,
                  'Department':department,
                   'Ext.':ext,
                  'Image':img,
                  
                
                   })
df1.shape
df1.to_csv('lhr_20l-1222.csv')
```

```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
options=Options()
options.add_argument("--headless=new")
options.add_argument("--window-size=1920,1200")
driver=webdriver.Chrome(options=options)
id_=[]
name=[]
email=[]
hec=[]
desig=[]
img=[]
ext=[]
dept=[]
high_edu=[]

count=0
for i in range(len(f2)-1):
    faculty2 = f2[i]
    try:
        id_.append(int(faculty2.find('ul',class_='social').li.a["href"].split('/')[-1]))
    except:
        id_.append(-1)
    print(faculty2.find('h3',class_='title').text)
    count=count+1
    print(count)
    name.append(faculty2.find('h3',class_='title').text)
    desig.append(faculty2.find('span',class_='post').text)
    email.append(faculty2.find('p').text)
    url2= "http://isb.nu.edu.pk/Faculty/Details/"+faculty2.find('ul',class_='social').li.a["href"].split('/')[-1]
    r2 = requests.get(url2) 
    print(r2) 
    soup3 = BeautifulSoup(r2.content, 'html.parser') 
    img.append("http://isb.nu.edu.pk"+soup3.find('div',class_="img-thumbnail").img["src"])
    a=[]
    a=soup3.find('li',class_='breadcrumb-item active').text.strip()
    if(a=='Computer Science'):
        dept.append('fsc')
    elif(a=='Electrical Engineering'):
        dept.append('ee')
    elif(a=='FAST School of Management'):
        dept.append('fsm')
    elif(a=='Science & Humanities'):
        dept.append('ss')
    else:
        dept.append('none')
    try:
        if(soup3.find('div',class_='profile-info').findAll('p')[1].find('label').text):
              hec.append("True")
    except:
        hec.append('none')
    try:
        ext.append(soup3.find('div',class_='profile-info').findAll('p')[-1].find('span').text.split(':')[-1])
    except:
        ext.append('none')
    try:
        driver.get(url2)
        educ=driver.find_element(By.CLASS_NAME,"tab-pane")
        high_edu.append(educ.text.split('\n')[2])
    except:
        high_edu.append('none')
        
        
driver.quit()
   
```

```python
df2 = pd.DataFrame({
                  'ID':id_,
                  'Name':name,
                  'Designation':desig,
                  'HEC approved':hec,
                 'Education':high_edu,
                  'Email':email,
                  'Department':dept,
                   'Ext.':ext,
                  'Image':img,
                     
                 })
df2.shape

df2.to_csv('isl_20l-1222.csv')
```

# FAST karachi

```python
URL_k= ["https://khi.nu.edu.pk/faculty-php/","https://khi.nu.edu.pk/department-of-artificial-intelligence/",
       "https://khi.nu.edu.pk/department-of-software-engineering/",
       "https://khi.nu.edu.pk/department-of-cyber-security/","https://khi.nu.edu.pk/department-of-electrical-engineering/",
    "https://khi.nu.edu.pk/department-of-management-sciences/","https://khi.nu.edu.pk/department-of-sciences-humanities/"]
id_=[]
name=[]
email=[]
desig=[]
img=[]
higher_edu=[]
ext=[]
department=[]
hec=[]
count=0
for e in range(len(URL_k)):
    rk = requests.get(URL_k[e])  
    print(rk) 
    soup4 = BeautifulSoup(rk.content, 'html.parser')
    # print(soup.prettify())
    fk = soup4.findAll('div', class_='gdlr-core-personnel-list gdlr-core-outer-frame-element clearfix')
    len(fk)
    for i in range(len(fk)):
        facultyk = fk[i]
        print(facultyk.find('h3', class_='gdlr-core-personnel-list-title').text.split(',')[0])
        count=count+1
        print(count)
        name.append(facultyk.find('h3', class_='gdlr-core-personnel-list-title').text.split(',')[0])
        desig.append(facultyk.find('div', class_='gdlr-core-personnel-list-position').text)
        try:
            img.append(facultyk.find('a').img["src"])
        except:
            img.append('none')   
        email.append(facultyk.find('div',class_="kingster-personnel-info-list kingster-type-email").text)
        if(e==0):
               department.append('fsc')
        elif(e==1):
              department.append('ai')
        elif(e==2):
               department.append('se')
        elif(e==3):
               department.append('cs')
        elif(e==4):
               department.append('ee')
        elif(e==5):
               department.append('fsm')
        elif(e==6):
               department.append('ss')
        else:
            department.append("none")
        try:
            ext.append(facultyk.find('div',class_="kingster-personnel-info-list kingster-type-phone").text)
        except:
             ext.append('none')
        url1k= facultyk.find('a')['href']
        print(url1k)
        r1k = requests.get(url1k) 
        print(r1k) 
        soupk = BeautifulSoup(r1k.content, 'html.parser')
        try:
                print(soupk.find('script',id='burst-js-extra').text.split(',')[1].split(':')[1])
                id_.append(soupk.find('script',id='burst-js-extra').text.split(',')[1].split(':')[1])
        except:
                id_.append(-1)
        faculty1k=soupk.find('div', class_="gdlr-core-pbf-column gdlr-core-column-40")
        try:
            if(faculty1k.find('span', class_='gdlr-core-title-item-caption gdlr-core-info-font gdlr-core-skin-caption').text.split('\n')[2]):
                 hec.append("True")
            else:
                hec.append("none")
        except:
            hec.append('none')
        try:
            higher_edu.append(faculty1k.find('div',class_='gdlr-core-icon-list-content-wrap').find('span',class_='gdlr-core-icon-list-content').text)
        except:
            higher_edu.append('none')
```

```python
df3 = pd.DataFrame({
                  'ID':id_,
                  'Name':name,
                  'Designation':desig,
                  'HEC approved':hec,
                 'Education':higher_edu,
                  'Email':email,
                  'Department':department,
                   'Ext.':ext,
                  'Image':img,
                
                   })
df3.to_csv('khi_20l-1222.csv')
```
